```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('2ND EDA - RMSE.csv')
df=df.iloc[-1:,1:]
df=df.transpose()
```

```python
df
```

```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]+0.1, y[i], ha = 'center')
if __name__ == '__main__':
    plt.figure(figsize=(12,6))
    x=df.index
    y=df[17]
    plt.bar(x,y,fill=False, hatch='/////')

    plt.xticks(rotation=45);
    addlabels(x, y)
    plt.title("RMSE Friedman Test (N*N) Rank",fontsize = 12,weight='bold' )
    plt.xlabel("Models",fontsize = 12,weight='bold')
    plt.ylabel("Ranks",fontsize = 12,weight='bold')
plt.savefig('Rank_Plot/RMSE Friedman Test (N*N) Rank',bbox_inches ="tight",pad_inches = 0.1,transparent = True,facecolor ="w",edgecolor ='w',orientation ='landscape')
```
