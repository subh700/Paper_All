```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
import time
#from pycaret.time_series import TSForecastingExperiment
import statsmodels.api as sm
```

```python
import glob
all_f= glob.glob('DataSets/'+'*/'+'*/'+'*_INDEX_*.csv')
all_f
li=[]
for file_name in all_f:
    df=pd.read_csv(file_name)
    li.append(df)
```

```python
df1=pd.DataFrame()
i=0
for i in range(0,17):
    df=li[i]
    v=all_f[i].split(r"/")
    D_n=v[2]
    i=i+1
    df1[D_n]=df.head(1).iloc[:,0:1]
```

```python
df1
```

```python
df2=pd.DataFrame()
i=0
for i in range(0,17):
    df=li[i]
    v=all_f[i].split(r"/")
    D_n=v[2]
    i=i+1
    df2[D_n]=df.tail(1).iloc[:,0:1].reset_index().iloc[:,1:2]
```

```python
df2
```

```python
frame=[df1,df2]
df3=pd.concat(frame)
df3=df3.reset_index()
df3=df3.drop('index',axis=1)
df3=df3.transpose()
df3=df3.rename(columns={0:'Fast_Day',1:'Last_Day'})
df3.to_csv('King.csv')
```
