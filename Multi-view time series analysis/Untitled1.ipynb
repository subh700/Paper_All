```python
import matplotlib.patches as patches

fig, ax = plt.subplots(4, 5, figsize=(18, 15), layout='constrained')
n = 24  # number of data points

row = 0  # row number
col = 0  # column number

for d in range(4):
    ax[row, col].plot(li[d].iloc[:,-11:].iloc[:n,0:1], label=li[d].iloc[:,-11:].iloc[:n,0:1].columns[0]+'_Data')
    ax[row, col].plot(li[d].iloc[:,-11:].iloc[:n,1:2], label=li[d].iloc[:,-11:].iloc[:n,1:2].columns[0].split(r'_')[0])
    ax[row, col].plot(li[d].iloc[:,-11:].iloc[:n,2:3], label=li[d].iloc[:,-11:].iloc[:n,2:3].columns[0].split(r'_')[0])
    ax[row, col].legend()
    ax[row, col].set_yticklabels([])  # Hide y-tick labels for this particular subplot

    # Add rectangle patch to the first subplot in each row
    rect = patches.Rectangle((1, 1), 1, 1, facecolor='red', alpha=1)
    ax[row, col].add_patch(rect)

    ax[row, col+1].plot(li[d].iloc[:,-11:].iloc[:n,0:1], label=li[d].iloc[:,-11:].iloc[:n,0:1].columns[0]+'_Data')
    ax[row, col+1].plot(li[d].iloc[:,-11:].iloc[:n,3:4], label=li[d].iloc[:,-11:].iloc[:n,3:4].columns[0].split(r'_')[0])
    ax[row, col+1].plot(li[d].iloc[:,-11:].iloc[:n,4:5], label=li[d].iloc[:,-11:].iloc[:n,4:5].columns[0].split(r'_')[0])
    ax[row, col+1].legend()

    ax[row, col+2].plot(li[d].iloc[:,-11:].iloc[:n,0:1], label=li[d].iloc[:,-11:].iloc[:n,0:1].columns[0]+'_Data')
    ax[row, col+2].plot(li[d].iloc[:,-11:].iloc[:n,5:6], label=li[d].iloc[:,-11:].iloc[:n,5:6].columns[0].split(r'_')[0])
    ax[row, col+2].plot(li[d].iloc[:,-11:].iloc[:n,6:7], label=li[d].iloc[:,-11:].iloc[:n,6:7].columns[0].split(r'_')[0])
    ax[row, col+2].legend()

    ax[row, col+3].plot(li[d].iloc[:,-11:].iloc[:n,0:1], label=li[d].iloc[:,-11:].iloc[:n,0:1].columns[0]+'_Data')
    ax[row, col+3].plot(li[d].iloc[:,-11:].iloc[:n,7:8], label=li[d].iloc[:,-11:].iloc[:n,7:8].columns[0].split(r'_')[0])
    ax[row, col+3].plot(li[d].iloc[:,-11:].iloc[:n,8:9], label=li[d].iloc[:,-11:].iloc[:n,8:9].columns[0].split(r'_')[0])
    ax[row, col+3].legend()

    ax[row, col+4].plot(li[d].iloc[:,-11:].iloc[:n,0:1], label=li[d].iloc[:,-11:].iloc[:n,0:1].columns[0]+'_Data')
    ax[row, col+4].plot(li[d].iloc[:,-11:].iloc[:n,9:10], label=li[d].iloc[:,-11:].iloc[:n,9:10].columns[0].split(r'_')[0])
    ax[row, col+4].plot(li[d].iloc[:,-11:].iloc[:n,10:11], label=li[d].iloc[:,-11:].iloc[:n,10:11].columns[0].split(r'_')[0])
    ax[row, col+4].legend()

    row += 1

plt.tight_layout()
plt.show()
```
