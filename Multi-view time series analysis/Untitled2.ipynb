# fredman test


```python
from scipy.stats import friedmanchisquare

ranks = ri.rank(ascending=False, axis=1).mean()
fvalue, pvalue = friedmanchisquare(*ri.values.T)
ranks = len(ranks) - ranks + 1

print("Modified Code: \n")
print("Ranks: \n", ranks)
print("F-value:", fvalue)
print("P-value:", pvalue)
```
