```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('RMSE.csv')
df
```

```python
rmse=df.iloc[:,:1]
rmse[df.iloc[:,2:3].columns[0]]=pd.DataFrame((((df.iloc[:,2:3].values-df.iloc[:,1:2].values)*100)/df.iloc[:,2:3].values))
rmse[df.iloc[:,4:5].columns[0]]=pd.DataFrame((((df.iloc[:,4:5].values-df.iloc[:,3:4].values)*100)/df.iloc[:,4:5].values))
rmse[df.iloc[:,6:7].columns[0]]=pd.DataFrame((((df.iloc[:,6:7].values-df.iloc[:,5:6].values)*100)/df.iloc[:,6:7].values))
rmse[df.iloc[:,8:9].columns[0]]=pd.DataFrame((((df.iloc[:,8:9].values-df.iloc[:,7:8].values)*100)/df.iloc[:,8:9].values))
rmse[df.iloc[:,10:11].columns[0]]=pd.DataFrame((((df.iloc[:,10:11].values-df.iloc[:,9:10].values)*100)/df.iloc[:,10:11].values))
rmse.to_csv('imp_rmse.csv')
```

```python
df=pd.read_csv('MAPE.csv')
df
```
