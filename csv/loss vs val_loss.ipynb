```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import textwrap
import os
```

```python
import glob
all_f= glob.glob('*/*_LOSS.csv')
all_f
```

```python
li=[]
for file_name in all_f:
    df=pd.read_csv(file_name)
    li.append(df)
```

```python
fig, axs = plt.subplots(6, 5, figsize=(22, 12), constrained_layout=True)
fig.subplots_adjust(right=0.8)  # Add space for legend
i=0
loss_df_list = []
name = []
v_loss = []
for dd in range(0, 28, 2):
    a = dd
    d_n = all_f[a].split('/')
    dn = d_n[0]
    if all_f[a].split('_')[-2] == 'ALL':
        loss_df = li[a]
        val_loss_df = li[a+1]
    else:
        loss_df = li[a+1]
        val_loss_df = li[a]

    v_loss.append(val_loss_df)
    loss_df_list.append(loss_df)
    name.append(dn)

for mn in range(0, 6, 2):
    if i<14:
        for co in range(0, 5):
            if (co+i)<14:
                axs[mn, co].plot(loss_df_list[co+i].iloc[:, 1:2].dropna(), label=loss_df_list[co+i].iloc[:, 1:2].columns[0],linestyle='dashed')
                axs[mn, co].plot(loss_df_list[co+i].iloc[:, 2:3].dropna(), label=loss_df_list[co+i].iloc[:, 2:3].columns[0],linestyle='dashed')
                axs[mn, co].plot(loss_df_list[co+i].iloc[:, 3:4].dropna(), label=loss_df_list[co+i].iloc[:, 3:4].columns[0],linestyle='dashed')
                axs[mn, co].plot(loss_df_list[co+i].iloc[:, 4:5].dropna(), label=loss_df_list[co+i].iloc[:, 4:5].columns[0],linestyle='dashed')
                axs[mn, co].plot(loss_df_list[co+i].iloc[:, 5:6].dropna(), label=loss_df_list[co+i].iloc[:, 5:6].columns[0],linestyle='dashed')
                axs[mn, co].legend()
                axs[mn,co].set_title(name[co+i], fontsize=12)
                if co == 4:
                    axs[mn,co].set_ylabel('Loss',rotation=270,fontsize=12,va='center',linespacing=1.5)
                    axs[mn, co].yaxis.set_label_coords(1.05, .5)
                if co==3 and mn==4:
                    axs[mn,co].set_ylabel('Loss',rotation=270,fontsize=12,va='center',linespacing=1.5)
                    axs[mn,co].yaxis.set_label_coords(1.05, .5)
                if mn == 4:
                    axs[mn+1, co].set_xlabel('No. of Epocs', fontsize=12, weight='bold')
                #axs[mn+1, co].plot(v_loss[co+i].iloc[:, 1:6].dropna(), label=v_loss[co+i].columns[1:6])
                axs[mn+1, co].plot(v_loss[co+i].iloc[:, 1:2].dropna(), label=v_loss[co+i].iloc[:, 1:2].columns[0])
                axs[mn+1, co].plot(v_loss[co+i].iloc[:, 2:3].dropna(), label=v_loss[co+i].iloc[:, 2:3].columns[0])
                axs[mn+1, co].plot(v_loss[co+i].iloc[:, 3:4].dropna(), label=v_loss[co+i].iloc[:, 3:4].columns[0])
                axs[mn+1, co].plot(v_loss[co+i].iloc[:, 4:5].dropna(), label=v_loss[co+i].iloc[:, 4:5].columns[0])
                axs[mn+1, co].plot(v_loss[co+i].iloc[:, 5:6].dropna(), label=v_loss[co+i].iloc[:, 5:6].columns[0])
                axs[mn+1, co].legend()
                if co == 4:
                    axs[mn+1,co].set_ylabel('Val_Loss',rotation=270,fontsize=12,va='center',linespacing=1.5)
                    axs[mn+1, co].yaxis.set_label_coords(1.05, .5)
                if co==3 and mn==4:
                    axs[mn+1,co].set_ylabel('Val_Loss',rotation=270,fontsize=12,va='center',linespacing=1.5)
                    axs[mn+1,co].yaxis.set_label_coords(1.05, .5)
                if mn == 4:
                    axs[mn+1,co].set_xlabel('No. of Epocs', fontsize=12, weight='bold')
                axs[3,4].set_xlabel('No. of Epocs', fontsize=12, weight='bold')
                axs[4,4].set_axis_off()
                axs[5,4].set_axis_off()
                axs[mn,co].legend().set_visible(False)
                axs[mn+1,co].legend().set_visible(False)
        i +=5
    handles, labels = axs[mn,co].get_legend_handles_labels()
    fig.legend(handles, labels, ncol=5, bbox_to_anchor=(.4, 1.04), fontsize=12, fancybox=True, framealpha=1)
    handles, labels = axs[mn+1,co].get_legend_handles_labels()
    fig.legend(handles, labels, ncol=5, bbox_to_anchor=(1, 1.04), fontsize=12, fancybox=True, framealpha=1)
plt.savefig('LOSS/loss_vs_val-loss.png',bbox_inches ="tight",pad_inches = 0.1,transparent = False,facecolor ="w",edgecolor ='w',orientation ='landscape')
```
