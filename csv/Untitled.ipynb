```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import textwrap
import os
```

```python
import glob
all_f= glob.glob('*/*Test_Pred_Copy*')
all_f
```

```python
li=[]
for file_name in all_f:
    df=pd.read_csv(file_name)
    li.append(df)
```

```python
fig, axs = plt.subplots(5, 5, figsize=(22, 12), layout='constrained')
for row in range(5):
    
    vv=[0,1,2,3,4]
    df=li[vv[row]]
    for col,k in enumerate([2,3,4,5,6]):
        j=7
        p_val=k+j-1
        ax= axs[row,col]
        ax.scatter(df.iloc[:, j-1:j],df.iloc[:,k-1:k],color='red',label='Normal')
        #ax.plot([0,df.iloc[:,j-1:j].max()],[0,df.iloc[:,k-1:k].max()],color='blue',label='Actual')
        ax.scatter(df.iloc[:, j-1:j],df.iloc[:,j+k-2:j+k-1],color='green',label='GAB')
        #ax.plot([0,df.iloc[:,j-1:j].max()],[0,df.iloc[:,j+k-2:j+k-1].max()],color='black',label='Actual Reference Line')
       # vvv=[df.iloc[:,k-1:k].max(),df.iloc[:,j+k-2:j+k-1].max()]
       # print(vvv)
       # print(df.iloc[:,k-1:k].max().values > df.iloc[:,j+k-2:j+k-1].max().values)
        if df.iloc[:,k-1:k].max().values > df.iloc[:,j+k-2:j+k-1].max().values:
            ax.plot([0,df.iloc[:,j-1:j].max()],[0,df.iloc[:,k-1:k].max()],color='black',label='Actual Reference Line')
        elif df.iloc[:,k-1:k].max().values == df.iloc[:,j+k-2:j+k-1].max().values:
            ax.plot([0,df.iloc[:,j-1:j].max()],[0,df.iloc[:,k-1:k].max()],color='blue',label='Actual')
        else:
            ax.plot([0,df.iloc[:,j-1:j].max()],[0,df.iloc[:,j+k-2:j+k-1].max()],color='black',label='Actual Reference Line')
        #ax.plot([0,df.iloc[:,j-1:j].max()],[0,vvv],color='blue',label='Actual')
        ax.legend(fancybox=True, framealpha=0.1)
        
        if row == 4:
            ax.set_xlabel('Actual', fontsize=12, weight='bold')
        
        if col == 0:
            ax.set_ylabel('Prediction', fontsize=12, weight='bold')
        if col == 4:
            ylabel_text = '\n'.join(textwrap.wrap(all_f[vv[row]].split(r"/")[0], width=17))
            ax.set_ylabel(ylabel_text, rotation=270, fontsize=12, va='center', linespacing=1.5)
            ax.yaxis.set_label_coords(1.1, .5)
        
        #ax.plot([0,df.iloc[:,j-1:j].max()],[0,df.iloc[:,j:j+1].max()],color='black',label='Actual Reference Line')
```
