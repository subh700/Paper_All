```python
for j in range(14):
    d_n=all_f[j].split(r'/')
    dn=d_n[0]
    li=[]
    for file_name in all_f:
        df=pd.read_csv(file_name)
        li.append(df)
    df=li[j].iloc[:,-5:]
    df.head(5)
    ff=pd.DataFrame()
    for i in range(5):
        model=['BiLSTM','LSTM','GRU','CNN','RNN']
        m=model[i]
        mo=m
        if df.iloc[:,0:1].columns[0] == 'models':
            run='models'
        elif df.iloc[:,0:1].columns[0] == 'Models':
            run='Models'
        else:
            run='1'
        mo = df.loc[df[run].isin([mo])] 
        mo=mo.reset_index()
        mo=mo.drop(columns=[run,'index'])
        v=mo.mean(axis=0)
        #ff=pd.DataFrame()
        ff[m]=v
        s=mo.std(axis=0)
        ff[m+'-std']=s
    ff=ff.T
    ff.to_csv('mean/'+dn+'_mean+std_r.csv')
ff
```
