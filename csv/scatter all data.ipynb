```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import textwrap
import os
```

```python
import glob
all_f= glob.glob('*/*Test_Pred_Copy*')
all_f
```

```python
li=[]
for file_name in all_f:
    df=pd.read_csv(file_name)
    li.append(df)
```

```python
fig, axs = plt.subplots(5, 5, figsize=(22, 12), layout='constrained')
for row in range(5):
    
    vv=[0,1,2,3,4]
    df=li[vv[row]]
    for col,k in enumerate([2,3,4,5,6]):
        j=7
        p_val=k+j-1
        ax= axs[row,col]
        ax.scatter(df.iloc[:, j-1:j],df.iloc[:,k-1:k],color= "#00000F",marker='^',label='Traditional')
        ax.scatter(df.iloc[:, j-1:j],df.iloc[:,j+k-2:j+k-1],color='#737373',label='GR2')
        
        if df.iloc[:,k-1:k].max().values > df.iloc[:,j+k-2:j+k-1].max().values:
            ax.plot([df.iloc[:,j-1:j].min(),df.iloc[:,j-1:j].max()],[df.iloc[:,k-1:k].min(),df.iloc[:,k-1:k].max()],color='black',label='Reference Line')
        elif df.iloc[:,k-1:k].max().values == df.iloc[:,j+k-2:j+k-1].max().values:
            ax.plot([df.iloc[:,j-1:j].min(),df.iloc[:,j-1:j].max()],[df.iloc[:,k-1:k].min(),df.iloc[:,k-1:k].max()],color='blue',label='Actual')
        else:
            ax.plot([df.iloc[:,j-1:j].min(),df.iloc[:,j-1:j].max()],[df.iloc[:,j+k-2:j+k-1].min(),df.iloc[:,j+k-2:j+k-1].max()],color='black',label='Reference Line')
        
        if row == 4:
            ax.set_xlabel('Orignal', fontsize=12, weight='bold')
        
        if col == 0:
            ax.set_ylabel('Prediction', fontsize=12, weight='bold')
        if col == 4:
            ylabel_text = '\n'.join(textwrap.wrap(all_f[vv[row]].split("\\")[0].split('_data')[0], width=17))
            ax.set_ylabel(ylabel_text, rotation=270, fontsize=12, va='center', linespacing=1.5)
            ax.yaxis.set_label_coords(1.1, .5)
        if row == 0:
            ax.set_title(df.iloc[:, k-1:k].columns[0], fontsize=12)
        # Hide the legend for subplots
        ax.legend().set_visible(False)
        
# Add a common legend outside of the subplots
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels,  ncol=5, bbox_to_anchor=(0.99, 1.04), fontsize=12, fancybox=True, framealpha=1)
# Full img titel
# plt.suptitle('scatter GAB V/s Normal',fontsize = 16,weight='bold')
# Save png
#os.mkdir('scatter')
plt.savefig('scatter/a_p_s_Gab_0 to 4.jpeg',dpi=400,bbox_inches ="tight",pad_inches = 0.1,transparent = False,facecolor ="w",edgecolor ='w',orientation ='landscape')
```
