```python
# Loop over range of 28
for dd in range(0,28,2):
    a = dd
    d_n = all_f[a].split(r'/')
    dn = d_n[0]
    
    # Get loss and validation loss dataframes
    if all_f[a].split(r'/')[0] == all_f[a+1].split(r'/')[0]:
        if all_f[a].split(r'_')[-2] == 'ALL':
            loss_df = li[a]
        elif all_f[a].split(r'_')[-2] == 'VAL':
            val_loss_df = li[a]
        if all_f[a+1].split(r'_')[-2]=='VAL':
            val_loss_df = li[a+1]
        elif all_f[a+1].split(r'_')[-2] == 'ALL':
            loss_df = li[a+1]
    else:
        if all_f[a].split(r'_')[-2] == 'ALL':
            loss_df = li[a]
        elif all_f[a].split(r'_')[-2] == 'VAL':
            val_loss_df = li[a]
        if all_f[a-1].split(r'_')[-2]=='VAL':
            val_loss_df = li[a-1]
        elif all_f[a-1].split(r'_')[-2] == 'ALL':
            loss_df = li[a-1]
    
    # Plot and save loss and validation loss plots
    for i in range(1, 6):
        plt.plot(val_loss_df.iloc[:,i], color='red', label='v_loss')
        plt.plot(loss_df.iloc[:,i], label='loss')
        plt.legend(loc='best')
        plt.savefig(f'su/{dn}_{loss_df.iloc[:,i-1:i].columns[0]}.png', bbox_inches='tight', pad_inches=0.1, transparent=False, facecolor='w', edgecolor='w', orientation='landscape')
        plt.cla()
```
