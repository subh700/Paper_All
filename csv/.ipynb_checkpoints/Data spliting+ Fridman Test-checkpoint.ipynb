```python
import pandas as pd
```

```python
import glob
all_f= glob.glob('*_Results.csv')
all_f
```

```python
fn=4
d_n=all_f[fn].split(r'.')
dn=d_n[0]+d_n[1]
dn=dn.split(r'_')
dn=dn[0]+'_'+dn[1]+'_'+dn[2]
dn
```

```python
li=[]
for file_name in all_f:
    df=pd.read_csv(file_name)
    li.append(df)
```

```python
df=li[fn]
df
```

```python
m='BiLSTM' #model
find='models' #find models in columns
ind='iteration_No' #index
mo=m
mo = df.loc[df[find].isin([mo])] 
mo=mo.reset_index()
mo=mo.drop(columns=['Unnamed: 0',find,'index'])
mo=mo.set_index(ind)
v=mo.mean(axis=0)
ff=pd.DataFrame()
ff[m]=v
ff
```
