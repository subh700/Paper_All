```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
li=[]
for file_name in all_f:
    df=pd.read_csv(file_name)
    df=df.iloc[:,-5:-2]
    if df.iloc[:,0:1].columns[0] == '1':
        BiLSTM = df[df['1'] == 'BiLSTM']
        BiLSTM=BiLSTM.reset_index()
        LSTM= df[df['1'] == 'LSTM']
        LSTM=LSTM.reset_index()
        GRU = df[df['1'] == 'GRU']
        GRU=GRU.reset_index()
        CNN= df[df['1'] == 'CNN']
        CNN=CNN.reset_index()
        RNN= df[df['1'] == 'RNN']
        RNN=RNN.reset_index()
    elif df.iloc[:,0:1].columns[0] == 'Models':
        BiLSTM = df[df['Models'] == 'BiLSTM']
        BiLSTM=BiLSTM.reset_index()
        LSTM= df[df['Models'] == 'LSTM']
        LSTM=LSTM.reset_index()
        GRU = df[df['Models'] == 'GRU']
        GRU=GRU.reset_index()
        CNN= df[df['Models'] == 'CNN']
        CNN=CNN.reset_index()
        RNN= df[df['Models'] == 'RNN']
        RNN=RNN.reset_index()
    elif df.iloc[:,0:1].columns[0] == 'models':
        BiLSTM = df[df['models'] == 'BiLSTM']
        BiLSTM=BiLSTM.reset_index()
        LSTM= df[df['models'] == 'LSTM']
        LSTM=LSTM.reset_index()
        GRU = df[df['models'] == 'GRU']
        GRU=GRU.reset_index()
        CNN= df[df['models'] == 'CNN']
        CNN=CNN.reset_index()
        RNN= df[df['models'] == 'RNN']
        RNN=RNN.reset_index()

        
    df1=pd.DataFrame()
    df1['BiLSTM-Tred_RMSE']=BiLSTM.iloc[:,-1:]
    df1['BiLSTM-GAB_RMSE']=BiLSTM.iloc[:,-2:-1]
    df1['LSTM-Tred_RMSE']=LSTM.iloc[:,-1:]
    df1['LSTM-GAB_RMSE']=LSTM.iloc[:,-2:-1]
    df1['GRU-Tred_RMSE']=GRU.iloc[:,-1:]
    df1['GRU-GAB_RMSE']=GRU.iloc[:,-2:-1]
    df1['CNN-Tred_RMSE']=CNN.iloc[:,-1:]
    df1['CNN-GAB_RMSE']=CNN.iloc[:,-2:-1]
    df1['RNN-Tred_RMSE']=RNN.iloc[:,-1:]
    df1['RNN-GAB_RMSE']=RNN.iloc[:,-2:-1]
    df1.to_csv('Boxplot/'+file_name.split('\\')[0]+'_rmse.csv')
    li.append(df1)
li[9]
```

```python
sns.set(rc={'figure.figsize':(8,8)})
li=[]
for j in range(10):
    d_n=all_f[j].split(r'/')
    dn=d_n[0]
    for file_name in all_f:
        df=pd.read_csv(file_name)
        li.append(df)
    df1=li[j].iloc[:,-5:-2]
    if df1.iloc[:,0:1].columns[0] == 'models':
        df=pd.melt(df1, id_vars=['models'], value_vars=['A_B_rmse', 'RMSE'])
        df=df.rename(columns={'models':'Models','value':'RMSE'})
        df['variable']=df['variable'].replace(['A_B_rmse','RMSE'],['GAB','Normal'])
    elif df1.iloc[:,0:1].columns[0] == '1':
        df=pd.melt(df1, id_vars=['1'], value_vars=['2', '3'])
        df=df.rename(columns={'1':'Models','value':'RMSE'})
        df['variable']=df['variable'].replace(['2','3'],['GAB','Normal'])
    else:
        df=pd.melt(df1, id_vars=['Models'], value_vars=['A_B_rmse', 'RMSE'])
        df=df.rename(columns={'Models':'Models','value':'RMSE'})
        df['variable']=df['variable'].replace(['A_B_rmse','RMSE'],['GAB','Normal'])
    ax = sns.boxplot(x="Models", y="RMSE", hue="variable", data=df)
    plt.legend(loc='best')
    plt.savefig('Boxplot/rmse/'+dn+'_rmse.png')
    plt.cla()
```
